```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#using the data set in the example
tips = sns.load_dataset("tips")
print(tips.head(n=4))
```

```python
menlunchtip = tips.loc[(tips['sex']=='Male')&(tips['time']=='Lunch'),'tip']
womenlunchtip = tips.loc[(tips['sex']=='Female')&(tips['time']=='Lunch'),'tip']
mendinnertip = tips.loc[(tips['sex']=='Male')&(tips['time']=='Dinner'),'tip']
womendinnertip = tips.loc[(tips['sex']=='Female')&(tips['time']=='Dinner'),'tip']
```

```python
#So far, all these graphs have focused on one continuous variable, 
#tips, and two categorical variables, gender and time. 
#Visualizing two continuous variables together involves a different type of plot.
```

```python
#Categorical variables contain a finite number of categories or distinct groups. 
#Categorical data might not have a logical order. 
#For example, categorical predictors include gender, material type, and payment method.
```

```python
#Continuous variables are numeric variables that have an infinite number of values between any two values. 
#A continuous variable can be numeric or date/time. 
#For example, the length of a part or the date and time a payment is received.
```

```python
plt.hist(mendinnertip, color='red', alpha=1, label='Dinner')
plt.hist(menlunchtip, color='black', alpha=1, label='Lunch')
plt.xlabel('Tip in USD')
plt.ylabel('# of People')
plt.title('Male Tips')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(womenlunchtip, color='black', alpha=.5, label='Women lunch')
plt.hist(womendinnertip, color='red', alpha=.5, label='Women dinner')
plt.xlabel('Tip in USD')
plt.ylabel('# of People')
plt.title('Female Tips')
plt.legend(loc='upper right')
plt.show()
```

```python
# Showing the  information in a seaborn facet grid.
sns.set(style="ticks")  #Setting the overall aesthetic

# Tell seaborn about the structure of our data.
g = sns.FacetGrid(tips, col="smoker", hue='sex')
# Name the plot type and the variable to be plotted using the structure.
g.map(plt.hist, "tip", color="steelblue",  lw=0)
# Moving the plots apart to make room for our titles.
plt.subplots_adjust(top=0.9)
# Making a more informative axis name.
g.set_axis_labels('Tip in USD')
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()

#use in data exploration in the beginning when trying to get some info on data
```

```python
g = sns.FacetGrid(tips, row="time", col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend()

#good way to plot continuous variables
#Bar graph wouldnt work for continuous variables
```

```python
g = sns.FacetGrid(tips, col="time", height=4, aspect=.5)
g.map(sns.barplot, "sex", "tip")
```

```python
sns.set(style="whitegrid")

g = sns.catplot(x="sex", y="tip", hue="time", data=tips,
                   height=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Tips")
g.set_xlabels("")
plt.title('Pointplot: Meal Tips by Time and Gender')
plt.show()
```

```python
sns.set(style="whitegrid")

ax = sns.boxplot(x='sex',y='tip',hue='time',data=tips,palette='pastel')  

plt.title('Plot 4: Meal Tips by Time and Gender')

sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tips')

plt.show()
```

```python
sns.set(style="white")

g = sns.lmplot(y='tip', 
               x='total_bill',
               hue='sex',
               data=tips, # Data.
               col='time',
               fit_reg=True,
               ci=False,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Tips")
g.set_xlabels("Total Bill")
plt.suptitle('Scatterplot Regression: Meal Tips by Total Bill, split by gender and time')
plt.subplots_adjust(top=0.9)
plt.show()
#"Bread and butter for a Data Scientist"
```
